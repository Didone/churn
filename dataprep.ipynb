```SQL
	USERNAME VARCHAR2(100),

	SALT VARCHAR2(100),
	MD5 VARCHAR2(100),
	SHA1 VARCHAR2(100),
	SHA256 VARCHAR2(100),
	CONSTRAINT LOGIN_PK PRIMARY KEY (UUID)
);
COMMENT ON TABLE CUSTOMERS.LOGIN IS 'Store Login credentials';
```

````python
%%sql
CREATE TABLE IF NOT EXISTS login
 USING CSV
SELECT _c34 `UUID`,
       _c16 `USERNAME`,
       _c17 `PASSWORD`,
       _c24 `REGISTERED`,
       _c18 `SALT`,
       _c19 `MD5`,
       _c20 `SHA1`,
       _c21 `SHA256`
  FROM csv.`workspace://churn/data/data`
````

```SQL
CREATE TABLE OT.USERS (
	UUID INTEGER NOT NULL,
	TITLE VARCHAR2(100),
	FIRST_NAME VARCHAR2(100),
	LAST_NAME VARCHAR2(100),
	GENDER VARCHAR2(100),
	EMAIL VARCHAR2(100),
	DOB DATE,
	PHONE VARCHAR2(100),
	CELL VARCHAR2(100),
	THUMBNAIL VARCHAR2(512),
	NAT VARCHAR2(100),
	CONSTRAINT USER_PK PRIMARY KEY (UUID),
	CONSTRAINT FK_LOGIN FOREIGN KEY (UUID) REFERENCES OT.LOGIN(UUID)
)
```
